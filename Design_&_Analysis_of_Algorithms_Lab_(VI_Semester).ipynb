T(n) = aT(n/b) + Θ(n<sup>k</sup>log<sup>p</sup>n)

1) if a > b<sup>k</sup>, Then T(n) = Θ(n<sup>log<sub>b</sub>a</sup>)

<pre style="font-size:16px; margin-top:0; margin-bottom:-20px"> (i) if p > -1, Then T(n) = Θ(n<sup>log<sub>b</sub>a</sup>log<sup>p+1</sup>n)

</pre>
3) if a < b<sup>k</sup>
<pre style="font-size:16px; margin-top:-20px"> (i) if p >= 0, Then T(n) = Θ(n<sup>k</sup>log<sup>p</sup>n)
(ii) if p < 0, Then T(n) = Θ(n<sup>k</sup>)

```python
## T(n) = 3T(n/2) + 2log(n)
def root(n, k):
    
    i = n;
    
    ## log(n)
    while (i > 1) :
        k = k + 3
        i = i / 2
    
    ## log(n)
    while (i > 1) :
        k = k + 3
        i = i / 2
    
    ## 3*T(n/2)
    if (n == 1):
        return 1
    else:
        return (root(n//2, 2 * k) + root(n//2, k) + root(n//2, k//2))
    
root(4, 2)
```

## MergeSort

```python
def mergeSort(A):

    n = len(A)
    if n==1: 
        return A
    mid = n//2 
    L = mergeSort(A[:mid])   # T(n/2)
    R = mergeSort(A[mid:])   # T(n/2)
    return merge(L,R)        # O(n)

# Merge Two Sequences
def merge(L,R):
    
    i = 0
    j = 0
    answer = []
    while i<len(L) and j<len(R):
        if L[i]<R[j]:
            answer.append(L[i])
            i += 1
        else:
            answer.append(R[j])
            j += 1
    if i<len(L):
        answer.extend(L[i:])
    if j<len(R):
        answer.extend(R[j:])
    return answer

if __name__ == "__main__":
    size = int(input("Enter the size of array: "))
    print("Elements of array: ")
    # arr = list(map(int, input().split()))[:size]
    arr = [int(input()) for _ in range(size)]
    print("original array:", arr)
    print("sorted array:", mergeSort(arr))
```

## Huffman Code data compression algorithm

```python
class PriorityQueue:   
    def __init__(self, arr):
        self.arr = self.heapify(arr)
        
    def __str__(self):
        return f"{self.arr}"
    
    def __len__(self):
        return len(self.arr)
    
    def __getitem__(self, index):
        return self.arr[index]

    @staticmethod
    def heapify(arr):
        n = len(arr)
        for i in range(n//2 - 1, -1, -1):
            PriorityQueue.min_heapify(arr, n, i)
        return arr

    @staticmethod
    def min_heapify(A, n, i):
        smallest = i
        l = 2 * i + 1
        r = 2 * i + 2
        if l < n and A[smallest] > A[l]:
            smallest = l
        if r < n and A[smallest] > A[r]:
            smallest = r
        if smallest != i:
            A[i], A[smallest] = A[smallest], A[i]
            PriorityQueue.min_heapify(A, n, smallest)

    def push(self, item):
        self.arr.append(item)
        pos = len(self.arr) - 1
        while self.arr[pos] < self.arr[pos//2]:
            self.arr[pos], self.arr[pos//2] = self.arr[pos//2], self.arr[pos]
            pos = pos//2
        return self.arr

    def pop(self):
        lastelt = self.arr.pop()
        if self.arr:
            returnitem = self.arr[0]
            self.arr[0] = lastelt
            self.min_heapify(self.arr, len(self.arr), 0)
            return returnitem
        return lastelt


class Huffman_Tree:
    
    class Node:
        def __init__(self, char, freq, left=None, right=None):
            self.char = char
            self.freq = freq
            self.left = left
            self.right = right
        
        def __repr__(self):
            if self.left is None and self.right is None:
                return f"{self.char}"
            elif self.left is None and self.right is not None:
                return f"{self.char}{self.right}"
            elif self.left is not None and self.right is None:
                return f"{self.left}{self.char}"
            else:
                return f"{self.left}{self.right}"
        
        def __len__(self):
            return len(f"{self.left}{self.right}")
        
        def __lt__(self, other):
            return self.freq < other.freq
        
    def __init__(self, string):
        self.string = string
        self.nodes = {}
        self.huffmanCode = self.encode()
        
    def __str__(self):
        def print2DTree_(root, space=1):
            if root.left is None and root.right is None:
                print(f"{' ':{space}s}|" + str(root) + "|")
                return
            if root is None: 
                return
            
            LEVEL_SPACE = len(root) + 3
            space += LEVEL_SPACE
            print2DTree_(root.right, space)
            # print() # neighbor space
            for i in range(LEVEL_SPACE, space): 
                print(end = " ")  
            print("|" + str(root) + "|<")
            print2DTree_(root.left, space)
            
        root = self.nodes[0]
        print(f"\n{' ':30s}Huffman Tree")
        print(f"{'-'*70}")
        print2DTree_(root)
        return ""
        
    def encode(self):
        # Calculating frequency
        for c in self.string: 
            self.nodes[c] = self.nodes.get(c, 0) + 1
        
        self.nodes = PriorityQueue([self.Node(k, v) for k, v in self.nodes.items()])

        while len(self.nodes) > 1:
            node1 = self.nodes.pop()  # heappop operation
            node2 = self.nodes.pop()  # heappop operation
            self.nodes.push(self.Node(None, node1.freq+node2.freq, node1, node2))  # heappush operation
        
        self.huffmanCode = self.huffman_code(self.nodes[0])
        return self.huffmanCode
    
    @staticmethod
    def huffman_code(node, left=True, code=''): 
        if node is None:
            return {node: code}
        
        if node.left is None and node.right is None:
            return {node.char: code}
        
        l = node.left
        r = node.right
    
        huffman_code = dict()
        huffman_code.update(Huffman_Tree.huffman_code(l, True, code + '0'))
        huffman_code.update(Huffman_Tree.huffman_code(r, False, code + '1'))
        return huffman_code
        
    def encoded_string(self):
        s = ''
        for c in self.string:
            s += self.huffmanCode.get(c)
        return s
    
    def decode(self, input_string=None):
        def isLeaf(root):
            return root.left is None and root.right is None
    
        def decode_(root, index, s):
            if root is None:
                return index

            # found a leaf node
            if isLeaf(root):
                print(root, end='')
                return index
 
            index = index + 1
            root = root.left if s[index] == '0' else root.right
            return decode_(root, index, s)
        
        root = self.nodes[0]
        s = self.encoded_string() if input_string is None else input_string
        
        # traverse the Huffman Tree again and decode the encoded string
        if root.left is None and root.right is None:
        # Special case: For input like a, aa, aaa, etc.
            while root.freq > 0:
                print(root, end='')
                root.freq = root.freq - 1 
        index = -1
        while index < len(s) - 1:
            index = decode_(root, index, s)
        return ''
```

## Fractional Knapsack

```python
class Item:
    def __init__(self, weight, value):
        self.weight = weight
        self.value = value

    def __lt__(self, other):
        return self.value//self.weight < other.value//other.weight

def fractional_knapsack(weight, value, capacity):

    items = [Item(weight[i], value[i]) for i in range(len(weight))]
    items.sort(reverse=True)

    totalValue = 0
    for item in items:
        current_weight = item.weight
        current_value = item.value
        if capacity - current_weight >= 0:
            capacity -= current_weight
            totalValue += current_value
        else:
            fraction = capacity / current_weight
            totalValue += current_value * fraction
            capacity = capacity - (current_weight * fraction)
            break
    return totalValue

weights  = [1, 4, 2, 3]
values   = [4, 4, 10, 12]
capacity = 8

print("Maximum value in Knapsack =", fractional_knapsack(weights, values, capacity))
```

## Minimum Spanning Tree (Kruskal's Algorithm)

```python
class DisjointSet:
    def __init__(self, num_nodes):
        self.parent = [i for i in range(num_nodes)]
        self.rank = [0] * num_nodes
        
    def findParent(self, node):
        if node == self.parent[node] :
            return node
        return self.findParent(self.parent[node])
    
    def union(self, edge):
        root1 = self.findParent(edge.src)
        root2 = self.findParent(edge.dst)
        
        if root1 != root2 :
            if self.rank[root1] < self.rank[root2] :
                self.parent[root1] = root2
                self.rank[root2] += 1
            else :
                self.parent[root2] = root1
                self.rank[root1] += 1
            return edge
        
class Edge :
    def __init__(self, src, destination, weight) :
        self.src = src
        self.dst = destination
        self.weight = weight
        
    def __lt__(self, other):
        return self.weight < other.weight
    
    def __repr__(self):
        return f"[{self.src}-{self.dst}]({self.weight})"
    
class Graph :
    def __init__(self, num_nodes, edgelist) :
        self.num_nodes = num_nodes
        self.edgelist = edgelist
        self.parent = []
        self.rank = []
        self.mst = []
    
    def KruskalMST (self) :
        self.edgelist.sort()
        ds = DisjointSet(self.num_nodes)
        
        # union find
        for edge in self.edgelist:
            edge_ = ds.union(edge)
            if edge_ is not None:
                self.mst.append(edge_)
                
        print ("Edge \tWeight")
        cost = 0
        for edge in self.mst :
            print(edge.src, "-", edge.dst, "\t", edge.weight)
            cost += edge.weight
        print("\nCost of minimum spanning tree : " +str(cost))

        
def createEdgeListFromAdjacencyMatrix(adj_mat):
    return [Edge(i, j, adj_mat[i][j]) for i in range(len(adj_mat)) \
            for j in range(i+1, len(adj_mat[i])) \
            if adj_mat[i][j] != 0] 
```

```python
class Edge :
    def __init__(self, src, destination, weight) :
        self.src = src
        self.dst = destination
        self.weight = weight
        
    def __lt__(self, other):
        return self.weight < other.weight
    
    def __repr__(self):
        return f"[{self.src}-{self.dst}]({self.weight})"
    
    def __str__(self):
        return f"[{self.src}-{self.dst}]({self.weight})"

class Vertex:
    def __init__(self, node):
        self.id = node
        self.adjacent = {}
        self.distance = 100000000     
        
    def add_neighbor(self, neighbor, weight=0):
        self.adjacent[neighbor] = weight
        
    def __getitem__(self, key):
        try:
            return self.adjacent[key]
        except:
            return None

    def __str__(self):
        return str({x:v for x,v in self.adjacent.items()})

    def __repr__(self):
        return str({x:v for x, v in self.adjacent.items()})
        
    def __lt__(self, other):
        return self.distance < other.distance
    
class Graph:
    def __init__(self):
        self.graph = {}

    def __iter__(self):
        return iter(self.graph.values())
    
    def add_vertex(self, node):
        for n in node:
            self.graph[n] = Vertex(n)
             
    def add_edge(self, edge: Edge):
        self.graph[edge.src].add_neighbor(edge.dst, edge.weight)
        self.graph[edge.dst].add_neighbor(edge.src, edge.weight)
        
    def get_vertices(self):
        return self.graph.keys()
    
    def get_outgoing_edges(self, vertex):
        connections = []
        for out_vertex in self.get_vertices():
            if self.graph[vertex][out_vertex] != None:
                connections.append(out_vertex)
        return connections
    
    def value(self, vertex1, vertex2):
        return self.graph[vertex1][vertex2]
        

def dijkstra_algorithm(graph, start_vertex, target_vertex=None):
    unvisited_vertices = list(graph.get_vertices())
 
    shortest_path = {vertex:10000000 for vertex in unvisited_vertices}
    shortest_path[start_vertex] = 0
    
    previous_vertices = {}
     
    while unvisited_vertices:
        current_min_vertex = None
        for vertex in unvisited_vertices:
            if current_min_vertex == None:
                current_min_vertex = vertex
            elif shortest_path[vertex] < shortest_path[current_min_vertex]:
                current_min_vertex = vertex
    
        neighbors = graph.get_outgoing_edges(current_min_vertex)
        for neighbor in neighbors:
            tentative_value = shortest_path[current_min_vertex] + graph.value(current_min_vertex, neighbor)
            if tentative_value < shortest_path[neighbor]:
                shortest_path[neighbor] = tentative_value
                previous_vertices[neighbor] = current_min_vertex
 
        unvisited_vertices.remove(current_min_vertex)

    if target_vertex:
        path = []
        
        vertex = target_vertex
        while vertex != start_vertex:
            path.append(vertex)
            vertex = previous_vertices[vertex]
 
        path.append(start_vertex)
    
        print("We found the following best path with a value of {}.".format(shortest_path[target_vertex]))
        print(" -> ".join(reversed(path)))
        
    else:
        return previous_vertices, shortest_path
        
g = Graph()
g.add_vertex(["BLP", "GD", "BHR", "BRB", "LKO", "GRP", "Kanpur", "LMP"])

e1, e2, e3 = Edge('BLP', 'GD', 42), Edge('BLP', 'BHR', 63), Edge("BLP", "GRP", 167)
e4, e5, e6 = Edge("GD", "LMP", 181), Edge('GD', 'BRB', 88), Edge('GD', 'BHR', 64)
e7, e8, e9 = Edge('GD', 'GRP', 172), Edge('LKO', 'GRP', 269), Edge('LKO', 'Kanpur', 100)
e10, e11, e12 = Edge('LMP', 'Kanpur', 225), Edge('BHR', 'Kanpur', 224), Edge('BHR', 'LMP', 121)
e13, e14, e15 = Edge("BHR", "BRB", 99), Edge('LKO', 'LMP', 130), Edge("LKO", "BRB", 28)

g.add_edge(e1), g.add_edge(e2), g.add_edge(e3), g.add_edge(e4), g.add_edge(e5)
g.add_edge(e6), g.add_edge(e7), g.add_edge(e8), g.add_edge(e9), g.add_edge(e10)
g.add_edge(e11), g.add_edge(e12), g.add_edge(e13), g.add_edge(e14), g.add_edge(e15)

dijkstra_algorithm(g, 'BLP', 'Kanpur')
```

#### using Binary Min-Heap

```python
import math

class BinaryHeap:   
    def __init__(self, arr=[]):
        self.arr = []
        
    def __str__(self):
        return f"{self.arr}"
    
    def __len__(self):
        return len(self.arr)
    
    def __getitem__(self, index):
        return self.arr[index]

    def heapify(self, arr):
        self.arr = arr
        n = len(self.arr)
        for i in range(n//2 - 1, -1, -1):
            self.min_heapify(self.arr, n, i)
        return self.arr

    def min_heapify(self, A, n, i):
        smallest = i
        l = 2 * i + 1
        r = 2 * i + 2
        if l < n and A[smallest] > A[l]:
            smallest = l
        if r < n and A[smallest] > A[r]:
            smallest = r
        if smallest != i:
            A[i], A[smallest] = A[smallest], A[i]
            self.min_heapify(A, n, smallest)

    def push(self, item):
        self.arr.append(item)
        pos = len(self.arr) - 1
        while self.arr[pos] < self.arr[pos//2]:
            self.arr[pos], self.arr[pos//2] = self.arr[pos//2], self.arr[pos]
            pos = pos//2
        return self.arr

    def pop(self):
        lastelt = self.arr.pop()
        if self.arr:
            returnitem = self.arr[0]
            self.arr[0] = lastelt
            self.min_heapify(self.arr, len(self.arr), 0)
            return returnitem
        return lastelt

class Edge :
    def __init__(self, src, destination, weight) :
        self.src = src
        self.dst = destination
        self.weight = weight
        
    def __lt__(self, other):
        return self.weight < other.weight
    
    def __repr__(self):
        return f"[{self.src}-{self.dst}]({self.weight})"
    
    def __str__(self):
        return f"[{self.src}-{self.dst}]({self.weight})"

class Vertex:
    def __init__(self, node):
        self.id = node
        self.adjacent = {}
        self.distance = math.inf
        self.visited = False
        
    def __getitem__(self, key):
        return self.adjacent[key]
        
    def __setitem__(self, key, value):
        self.adjacent[key] = value
        
    def __lt__(self, other):
        return self.distance < other.distance

    def __str__(self):
        return str({x.id for x in self.adjacent})
    
    def __repr__(self):
        return str({x.id for x in self.adjacent})
    
class Graph:
    def __init__(self):
        self.graph = {}

    def __iter__(self):
        return iter(self.graph.values())
    
    def __getitem__(self, key):
        return self.graph[key]
    
    def add_vertex(self, vertex):
        if type(vertex) is list:
            for n in vertex:
                self.graph[n] = Vertex(n)
        else:
            self.graph[vertex] = Vertex(vertex)
             
    def add_edge(self, edges:list):
        try:
            for edge in edges:
                if edge.src not in self.graph:
                    self.add_vertex(edge.src)
                if edge.dst not in self.graph:
                    self.add_vertex(edge.dst)
                    
                self.graph[edge.src][self.graph[edge.dst]] = edge.weight
                self.graph[edge.dst][self.graph[edge.src]] = edge.weight
        except TypeError:
            print(f'{edges} edges should be type of list')
        
    def get_vertex(self, n):
        if n in self.graph:
            return self.graph[n]
        else:
            return None
        
    def get_vertices(self):
        return self.graph.keys()
    
    def value(self, vertex1, vertex2):
        return self.graph[vertex1][vertex2]
    
    def erase(self):
        for v in self:
            v.visited = False
            v.distance = math.inf
        
def dijkstra_algorithm(graph, start_vertex, target_vertex=None, erase=False):    
    graph.get_vertex(start_vertex).distance = 0
    
    unvisited_queue = BinaryHeap()
    unvisited_queue.heapify([(v, v.distance) for v in graph])
    
    shortest_path = {vertex[0].id:math.inf for vertex in unvisited_queue}
    shortest_path[start_vertex] = 0
    
    previous_vertices = {}
    
    while len(unvisited_queue):
        uv = unvisited_queue.pop()
        current = uv[0]
        current.visited = True

        for next_ in current.adjacent:
            if next_.visited:
                continue
            new_dist = current.distance + current[next_]
            if new_dist < next_.distance:
                next_.distance = new_dist
                shortest_path[next_.id] = new_dist
                previous_vertices[next_.id] = current.id

        # Rebuild heap
        # 1. Pop every item
        while len(unvisited_queue):
            unvisited_queue.pop()
        # 2. Put all vertices not visited into the queue
        unvisited_queue = BinaryHeap()
        unvisited_queue.heapify([(v, v.distance) for v in graph if not v.visited])
        
    if erase:
        graph.erase()
     
    if target_vertex:
        path = []
        
        vertex = target_vertex
        while vertex != start_vertex:
            path.append(vertex)
            vertex = previous_vertices[graph.get_vertex(vertex).id]
 
        path.append(start_vertex)
    
        print(f"We found the following best path with a value of {shortest_path[target_vertex]}KM.")
        print(" -> ".join(reversed(path)))
        
    else:
        return previous_vertices, shortest_path
        
g = Graph()
g.add_vertex(["Gorakhpur", "Basti", "Siddharthanagar", "Balrampur", "Ayodhya", "Gonda", "Bahraich", 
              "Barabanki", "Lucknow", "Unnao", "Kanpur", "Sitapur", "Lakhimpur"])

e1, e2 = Edge('Balrampur', 'Gonda', 42), Edge('Balrampur', 'Bahraich', 63)
e3, e4 = Edge('Balrampur', 'Gorakhpur', 162), Edge('Balrampur', 'Siddharthanagar', 75)
e5, e6 = Edge('Balrampur', 'Basti', 103), Edge('Gonda', 'Barabanki', 88)
e7, e8 = Edge('Gonda', 'Siddharthanagar', 107), Edge('Gonda', 'Lakhimpur', 181)
e9, e10 = Edge('Gonda', 'Gorakhpur', 160), Edge('Gonda', 'Basti', 96)
e11, e12 = Edge('Gonda', 'Bahraich', 64), Edge('Gonda', 'Ayodhya', 50)
e13, e14 = Edge('Lucknow', 'Gorakhpur', 269), Edge('Lucknow', 'Unnao', 64)
e15, e16 = Edge("Lucknow", 'Barabanki', 28), Edge('Lucknow', 'Sitapur', 87)
e17, e18 = Edge('Siddharthanagar', 'Gorakhpur', 98), Edge('Siddharthanagar', 'Basti', 69)
e19, e20 = Edge('Ayodhya', 'Siddharthanagar', 108), Edge('Ayodhya', 'Barabanki', 109)
e21, e22 = Edge('Ayodhya', 'Basti', 69), Edge('Lakhimpur', 'Kanpur', 225),
e23, e24 = Edge('Lakhimpur', 'Barabanki', 131), Edge('Lakhimpur', 'Sitapur', 45)
e25, e26 = Edge('Bahraich', 'Lakhimpur', 121), Edge('Bahraich', 'Barabanki', 99)
e27, e28 = Edge('Bahraich', 'Kanpur', 224), Edge('Basti', 'Gorakhpur', 65)
e29, e30 = Edge('Unnao', 'Kanpur', 22), Edge('Unnao', 'Sitapur', 136)

g.add_edge([e1, e2, e3, e4, e5, e6, e7, e8, e9, e10, e11, e12, e13, e14, e15, e16, e17, e18, e19, e20])
g.add_edge([e21, e22, e23, e24, e25, e26, e27, e28, e29, e30])
dijkstra_algorithm(g, start_vertex='Balrampur', target_vertex='Kanpur')
```

```python
import math

class FibonacciHeap:
    class FibonacciTree:
        def __init__(self, value):
            self.value = value
            self.order = 0
            self.child = []
        
        def add_at_end(self, t):
            self.child.append(t)
            self.order = self.order + 1
        
    def __init__(self):
        self.least = None
        self.count = 0
        self.arr = []
        
    def __str__(self):
        return f"{[x.value for x in self.arr]}"
    
    def __len__(self):
        return len(self.arr)
    
    def __getitem__(self, index):
        return self.arr[index]

    def push(self, value):
        new_tree = self.FibonacciTree(value)
        self.arr.append(new_tree)
        if (self.least is None or value < self.least.value):
            self.least = new_tree
        self.count = self.count + 1
        return self.arr

    def get_min(self):
        if self.least is None:
            return None
        return self.least.value
    
    def pop(self):
        smallest = self.least
        if smallest is not None:
            for child in smallest.child:
                self.arr.append(child)
            self.arr.remove(smallest)
            if self.arr == []:
                self.least = None
            else:
                self.least = self.arr[0]
                self.heapify()
            self.count = self.count - 1
            return smallest.value
        
    def heapify(self, arr=[]):
        if arr != []:
            return [self.push(v) for v in arr]
        
        aux = math.frexp(self.count)[1] * [None]    
        while self.arr != []:
            x = self.arr[0]
            order = x.order
            self.arr.remove(x)
            while aux[order] is not None:
                y = aux[order]
                if x.value > y.value:
                    x, y = y, x
                x.add_at_end(y)
                aux[order] = None
                order += 1
            aux[order] = x
        self.least = None
        for k in aux:
            if k is not None:
                self.arr.append(k)
                if self.least is None or k.value < self.least.value:
                    self.least = k

class Edge :
    def __init__(self, src, destination, weight) :
        self.src = src
        self.dst = destination
        self.weight = weight
        
    def __lt__(self, other):
        return self.weight < other.weight
    
    def __repr__(self):
        return f"[{self.src}-{self.dst}]({self.weight})"
    
    def __str__(self):
        return f"[{self.src}-{self.dst}]({self.weight})"

class Vertex:
    def __init__(self, node):
        self.id = node
        self.adjacent = {}
        self.distance = 100000000 
        self.visited = False
        
    def __getitem__(self, key):
        return self.adjacent[key]
        
    def __setitem__(self, key, value):
        self.adjacent[key] = value
        
    def __lt__(self, other):
        return self.distance < other.distance

    def __str__(self):
        return str({x.id for x in self.adjacent})
    
    def __repr__(self):
        return str({x.id for x in self.adjacent})
    
class Graph:
    def __init__(self):
        self.graph = {}

    def __iter__(self):
        return iter(self.graph.values())
    
    def __getitem__(self, key):
        return self.graph[key]
    
    def add_vertex(self, vertex):
        if type(vertex) is list:
            for n in vertex:
                self.graph[n] = Vertex(n)
        else:
            self.graph[vertex] = Vertex(vertex)
             
    def add_edge(self, edges:list):
        try:
            for edge in edges:
                if edge.src not in self.graph:
                    self.add_vertex(edge.src)
                if edge.dst not in self.graph:
                    self.add_vertex(edge.dst)
                    
                self.graph[edge.src][self.graph[edge.dst]] = edge.weight
                self.graph[edge.dst][self.graph[edge.src]] = edge.weight
        except TypeError:
            print(f'{edges} edges should be type of list')

    def get_vertex(self, n):
        if n in self.graph:
            return self.graph[n]
        else:
            return None
    
    def value(self, vertex1, vertex2):
        return self.graph[vertex1][vertex2]
    
    def erase(self):
        for v in self:
            v.visited = False
            v.distance = math.inf
        
def dijkstra_algorithm(graph, start_vertex, target_vertex=None, erase=False):
    graph.get_vertex(start_vertex).distance = 0
    
    unvisited_queue = FibonacciHeap()
    unvisited_queue.heapify([(v, v.distance) for v in graph])
    
    shortest_path = {vertex.value[0].id:math.inf for vertex in unvisited_queue}
    shortest_path[start_vertex] = 0
    
    previous_vertices = {}
    
    while len(unvisited_queue):
        uv = unvisited_queue.pop()
        current = uv[0]
        current.visited = True

        for next_ in current.adjacent:
            if next_.visited:
                continue
            new_dist = current.distance + current[next_]
            
            if new_dist < next_.distance:
                next_.distance = new_dist
                shortest_path[next_.id] = new_dist
                previous_vertices[next_.id] = current.id
        while len(unvisited_queue):
            unvisited_queue.pop()
            
        unvisited_queue = FibonacciHeap()
        unvisited_queue.heapify([(v, v.distance) for v in graph if not v.visited])
        
    if erase:
        graph.erase()
    
    if target_vertex:
        path = []
        
        vertex = target_vertex
        while vertex != start_vertex:
            path.append(vertex)
            vertex = previous_vertices[vertex]
 
        path.append(start_vertex)
    
        print(f"We found the following best path with a value of {shortest_path[target_vertex]}KM.")
        print(" -> ".join(reversed(path)))
        
    else:
        return previous_vertices, shortest_path
        
g = Graph()
g.add_vertex(["Gorakhpur", "Basti", "Siddharthanagar", "Balrampur", "Ayodhya", "Gonda", "Bahraich", 
              "Barabanki", "Lucknow", "Unnao", "Kanpur", "Sitapur", "Lakhimpur"])

e1, e2 = Edge('Balrampur', 'Gonda', 42), Edge('Balrampur', 'Bahraich', 63)
e3, e4 = Edge('Balrampur', 'Gorakhpur', 162), Edge('Balrampur', 'Siddharthanagar', 75)
e5, e6 = Edge('Balrampur', 'Basti', 103), Edge('Gonda', 'Barabanki', 88)
e7, e8 = Edge('Gonda', 'Siddharthanagar', 107), Edge('Gonda', 'Lakhimpur', 181)
e9, e10 = Edge('Gonda', 'Gorakhpur', 160), Edge('Gonda', 'Basti', 96)
e11, e12 = Edge('Gonda', 'Bahraich', 64), Edge('Gonda', 'Ayodhya', 50)
e13, e14 = Edge('Lucknow', 'Gorakhpur', 269), Edge('Lucknow', 'Unnao', 64)
e15, e16 = Edge("Lucknow", 'Barabanki', 28), Edge('Lucknow', 'Sitapur', 87)
e17, e18 = Edge('Siddharthanagar', 'Gorakhpur', 98), Edge('Siddharthanagar', 'Basti', 69)
e19, e20 = Edge('Ayodhya', 'Siddharthanagar', 108), Edge('Ayodhya', 'Barabanki', 109)
e21, e22 = Edge('Ayodhya', 'Basti', 69), Edge('Lakhimpur', 'Kanpur', 225),
e23, e24 = Edge('Lakhimpur', 'Barabanki', 131), Edge('Lakhimpur', 'Sitapur', 45)
e25, e26 = Edge('Bahraich', 'Lakhimpur', 121), Edge('Bahraich', 'Barabanki', 99)
e27, e28 = Edge('Bahraich', 'Kanpur', 224), Edge('Basti', 'Gorakhpur', 65)
e29, e30 = Edge('Unnao', 'Kanpur', 22), Edge('Unnao', 'Sitapur', 136)

g.add_edge([e1, e2, e3, e4, e5, e6, e7, e8, e9, e10, e11, e12, e13, e14, e15, e16, e17, e18, e19, e20])
g.add_edge([e21, e22, e23, e24, e25, e26, e27, e28, e29, e30])

dijkstra_algorithm(g, start_vertex='Lakhimpur', target_vertex='Kanpur')
```

## Floyd Warshall Algorithm

```python
import math

def floyd_warshall(G, nV):
    for k in range(nV):
        for i in range(nV):
            for j in range(nV):
                G[i][j] = min(G[i][j], G[i][k] + G[k][j])
    return G

G = [[0, 3, math.inf, 5],
         [2, 0, math.inf, 4],
         [math.inf, 1, 0, math.inf],
         [math.inf, math.inf, 2, 0]]
floyd_warshall(G, 4)
```

## 0/1 Knapsack

```python
def ZeroOneKnapsack(weights, values, capacity, n):
    KNAPSACK = [0 for i in range(capacity+1)]
    for i in range(1, n+1):
        for j in range(capacity, weights[i-1]-1, -1):
            KNAPSACK[j] = max(values[i-1] + KNAPSACK[j- weights[i-1]], KNAPSACK[j])
    return KNAPSACK[capacity]
```

## DFS

### for connected graph

```python
def dfs(graph, visited, queue, node):
    if node in visited: return
    visited.append(node)
    queue.append(node)
    while queue:
        current_node = queue.pop(0)
        print(current_node, end=' ')
        for neighbor in graph[current_node]:
            dfs(graph, visited, queue, neighbor)

graph = {
    'A': ['B', 'C', 'D'],
    'B': ['E', 'F'],
    'C': ['F', 'G'],
    'D': ['H'],
    'E': ['H'],
    'F': ['G', 'D'],
    'G': ['D'],
    'H': []
    }

dfs(graph, [], [], 'A')
```

### for disconnnected Graph

```python
def dfs(graph, visited):
    def dfs_(graph, visited, node):
        visited.append(node)
        print(node, end=' ')
        for neighbor in graph[node]:
            if neighbor not in visited:
                dfs_(graph, visited, neighbor)

    for v in graph.keys():
        if v not in visited:
            dfs_(graph, visited, v)

if __name__ == '__main__':
    # disconnected graph
    graph = {
        'A': ['B', 'C'],
        'B': ['F'],
        'C': [],
        'D': ['A', 'H', 'G'],
        'E': ['H'],
        'F': ['C'],
        'G': ['E'],
        'H': []
    }
    
    dfs(graph, [])
```

## Strongly Connected Components (Kosaraju's Algorithm)

```python
from collections import defaultdict

class Graph:
    def __init__(self, vertex):
        self.V = vertex
        self.graph = defaultdict(list)

    def add_edge(self, s, d):
        self.graph[s].append(d)

    def dfs(self, d, visited_vertex):
        visited_vertex[d] = True
        print(d, end='')
        for i in self.graph[d]:
            if not visited_vertex[i]:
                self.dfs(i, visited_vertex)

    def fill_order(self, d, visited_vertex, stack):
        visited_vertex[d] = True
        for i in self.graph[d]:
            if not visited_vertex[i]:
                self.fill_order(i, visited_vertex, stack)
        stack = stack.append(d)

    def transpose(self):
        g = Graph(self.V)

        for i in self.graph:
            for j in self.graph[i]:
                g.add_edge(j, i)
        return g

    def scc(self):
        stack = []
        visited_vertex = [False] * (self.V)

        for i in range(self.V):
            if not visited_vertex[i]:
                self.fill_order(i, visited_vertex, stack)

        gr = self.transpose()

        visited_vertex = [False] * (self.V)

        while stack:
            i = stack.pop()
            if not visited_vertex[i]:
                gr.dfs(i, visited_vertex)
                print("")


g = Graph(8)
g.add_edge(0, 1)
g.add_edge(1, 2)
g.add_edge(2, 3)
g.add_edge(2, 4)
g.add_edge(3, 0)
g.add_edge(4, 5)
g.add_edge(5, 6)
g.add_edge(6, 4)
g.add_edge(6, 7)

print("Strongly Connected Components:")
g.scc()
```
